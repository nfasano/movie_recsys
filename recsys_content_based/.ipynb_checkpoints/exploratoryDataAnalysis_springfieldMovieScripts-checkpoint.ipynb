```python
# Load in python packages
import requests
from bs4 import BeautifulSoup

import re
import numpy as np
import pandas as pd

import datetime
import time

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation

import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)

```

```python
path_to_csv = 'SpringField_MovieScripts_2023_01_13_21hr_21min_28sec.csv' 
df = pd.read_csv(path_to_csv)
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Movie Year"] = df["Movie Title"].map(lambda x: x[-5:-1])
```

```python
df["Movie Year"].unique()
```

```python
df.loc[df["Movie Year"] == '0000',"Movie Year"] = '1986'
df.loc[df["Movie Year"] == '2050',"Movie Year"] = '2017'
df.loc[df["Movie Year"] == '0147',"Movie Year"] = '2014'
df.loc[df["Movie Year"] == 'arth',"Movie Year"] = '2021'
```

```python
df['Movie Year'] = df['Movie Year'].astype('int')
```

```python
a = sns.histplot(df,x="Movie Year",binwidth=1)
a.set(ylabel='Movie Count');
```

```python
dff = df.copy()
```

```python
dff.head()
```

```python
# remove punctuation
dff['Script Text'] = dff['Script Text'].str.replace(r'[^\w\s]+', '') # uses a regular expression \w == all words \s == all spaces [] == set of characters + == allow for mulitple occurences

```

```python
# make everything lowercase
dff['Script Text'] = dff['Script Text'].str.lower()
```

```python
# remove a set of stop words
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stopWordsList = stopwords.words('english')
```

```python
stopWordsList
# remove stop words
dff['Script Text'] = dff['Script Text'].apply([lambda x:" ".join([word for word in x.split() if word not in stopWordsList])])
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(dff['Script Text'])
vocab = vectorizer.vocabulary_
```

```python
vocab
```

```python
vocab['000425660042759default000000000000hi']
```

```python
X.shape
```

```python
np.sum(X[:,1164367])
```

```python
wordCountAllDocs = np.array(np.sum(X,axis=0))[0]
```

```python
indw = np.argsort(wordCountAllDocs)
indw = np.flip(indw)
indw
```

```python
wordCountAllDocs = wordCountAllDocs[indw]
```

```python
wordKeys = list(vocab.keys())
wordKeysSort = [wordKeys[list(vocab.values()).index(i)] for i in indw[0:1000]]
```

```python
wordCountAllDocs
```

```python
xx = np.linspace(0,np.size(wordCountAllDocs),np.size(wordCountAllDocs))
```

```python
numWords = 100

fig, ax = plt.subplots(figsize=(16, 8))

ax.bar(x=xx[0:numWords+1],height=wordCountAllDocs[0:numWords+1])
ax.set_ylim(0,np.max(wordCountAllDocs)+1)
ax.set_xlim(0,numWords)
ax.set_ylabel('Word Count')
ax.set_xlabel('Word ID')
ax.set_xticks(np.linspace(0,numWords,numWords+1));
ax.set_xticklabels(wordKeysSort[0:numWords+1])
plt.xticks(rotation=90);
```
